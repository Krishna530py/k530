# Dictionary inside dictionary

```python
d = {1:{'name': 'Sita', 'ntc': '9878884654', 'ncell': '9808278765'},
    2:{'name': 'Shyam', 'ntc': '9867667654', 'ncell': '9808776765'}}
print(d)
```

```python
info = {}
n = int(input("Enter n = "))
for i in range(1,n+1):
    name = input("Enter name = ")
    ntc_phone = input("Enter ntc phone = ")
    ncell_phone = input("Enter ncell phone = ")
    data = {'name':name,'ntc':ntc_phone,'ncell':ncell_phone}
    info[i] = data

    
print(info)
```
